```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

folder_path = '/content/sentence-classification/data/SPAADIA/'  # Update if different
rows = []

for filename in os.listdir(folder_path):
    if filename.endswith(".xml"):
        filepath = os.path.join(folder_path, filename)
        tree = ET.parse(filepath)
        root = tree.getroot()

        for sentence in root.findall(".//s"):
            # Reconstruct sentence text
            words = [w.text for w in sentence.findall(".//w") if w.text]
            sentence_text = ' '.join(words)

            # Determine sentence type from attributes
            sent_type = sentence.get("illoc") or sentence.get("mod") or sentence.get("type")

            # Normalize labels
            if sent_type:
                sent_type = sent_type.lower()
                if "decl" in sent_type or "assert" in sent_type:
                    label = "assertive"
                elif "inter" in sent_type or "ques" in sent_type:
                    label = "interrogative"
                elif "exclam" in sent_type:
                    label = "exclamatory"
                elif "command" in sent_type or "imper" in sent_type or "dir" in sent_type:
                    label = "imperative"
                else:
                    continue  # skip undefined types
            else:
                continue  # skip if no sentence type

            # Clean and save
            if sentence_text.strip():
                rows.append({"sentence": sentence_text.strip(), "label": label})

# Create DataFrame
df = pd.DataFrame(rows)
df.drop_duplicates(inplace=True)
```

```python
import os
import xml.etree.ElementTree as ET
import pandas as pd

folder_path = '/content/sentence-classification/data/SPAADIA/'  # adjust if needed
rows = []

def get_label(tag, attrib):
    tag = tag.lower()
    mode = attrib.get('mode', '').lower()
    sp_act = attrib.get('sp-act', '').lower()

    if "query" in mode or tag.startswith("q-"):
        return "interrogative"
    elif "exclam" in mode or sp_act == "exclaim":
        return "exclamatory"
    elif "intent" in mode or "command" in mode or "req" in sp_act:
        return "imperative"
    elif "decl" in mode or tag == "decl" or tag == "frag":
        return "assertive"
    else:
        return None

for file in os.listdir(folder_path):
    if file.endswith(".xml"):
        path = os.path.join(folder_path, file)
        try:
            tree = ET.parse(path)
            root = tree.getroot()
        except ET.ParseError:
            continue  # Skip malformed files

        for turn in root.findall(".//turn"):
            for elem in turn:
                label = get_label(elem.tag, elem.attrib)
                if label:
                    text = ''.join(elem.itertext()).strip()
                    if text:
                        rows.append({"sentence": text, "label": label})

# Create and save DataFrame
df = pd.DataFrame(rows).drop_duplicates()
```

```python
# 2. Load and preprocess your CSV dataset
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Replace with your actual dataset path
df1 = df1.dropna()
df1 = df1.sample(frac=1, random_state=42).reset_index(drop=True)  # shuffle

# Encode labels
label_encoder = LabelEncoder()
df1['label_id'] = label_encoder.fit_transform(df1['label'])
label2id = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
id2label = {v: k for k, v in label2id.items()}
```

```python
# 3. Convert to Hugging Face Dataset
from datasets import Dataset

dataset = Dataset.from_pandas(df1[['sentence', 'label_id']])
```

```python
# 4. Tokenize sentences using bert-base-uncased
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

def tokenize(example):
    return tokenizer(example['sentence'], truncation=True, padding='max_length', max_length=128)

dataset = dataset.map(tokenize)
```

```python
# 5. Train-test split
dataset = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = dataset['train']
test_dataset = dataset['test']
```

```python
# Fix: Convert keys to native Python int
id2label = {int(k): v for k, v in id2label.items()}
label2id = {v: int(k) for k, v in id2label.items()}
```

```python
# 6. Load BERT model for classification
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(
    'bert-base-uncased',
    num_labels=3,
    id2label=id2label,
    label2id=label2id
)
```

```python
# 7. Set up training arguments
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=4,
    per_device_train_batch_size=8,
    evaluation_strategy='epoch',
    save_strategy='epoch',
    logging_dir='./logs',
    logging_steps=10,
    load_best_model_at_end=True
)
```

```python
# 8. Train the model
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    tokenizer=tokenizer
)

trainer.train()
```

```python
#import transformers
print(transformers.__version__)  # should be 4.40.1
```

```python
# 9. Evaluate the model
metrics = trainer.evaluate()
print(metrics)
```

```python
# 10. Inference on new sentence
def classify_sentence(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    outputs = model(**inputs)
    pred_id = outputs.logits.argmax(dim=1).item()
    return id2label[pred_id]

# Example
print(classify_sentence("Please help me with this!"))  # exclamatory?
```
