```python
def get_comments(video_id):
    comments = []
    next_page_token = None

    while True:
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            maxResults=100,
            pageToken=next_page_token,
            textFormat='plainText'
        ).execute()

        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append({
                'author': comment['authorDisplayName'],
                'text': comment['textDisplay'],
                'likes': comment['likeCount'],
                'published_at': comment['publishedAt']
            })

        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break

    return pd.DataFrame(comments)
```

```python
from googleapiclient.discovery import build
from indic_transliteration.sanscript import transliterate, ITRANS, DEVANAGARI
from langdetect import detect
import regex as re
import pandas as pd
import numpy as np
import emoji
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('dhruv_rathee.csv')
df1 = df[['text', 'likes', 'Date', 'Time']]

df1['text'] = df1['text'].replace(r'^(nan|None)$', '', regex=True)
# Replace only real NaN (Not a Number), not the string "nan"
df1['text'] = df1['text'].fillna('')         # fills only actual NaN values
# Convert all values to string after that
df1['text'] = df1['text'].astype(str)

# Remove rows where 'text' is an empty string
df1 = df1[df1['text'].str.strip() != '']
df1.reset_index(drop=True, inplace=True)
df1 = df1.sort_values(by='likes', ascending=False, ignore_index=True)

df1['text'] = df1['text'].str.lower()

df1['text'] = df1['text'].apply(emoji.demojize)

def remove_HTML_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text) # Convert to string to handle potential non-string values
df1['text'] = df1['text'].apply(remove_HTML_tags) # Assign the result back to the column

def remove_url(text):
	pattern = re.compile(r'https://\S+|www\.\S+')
	return pattern.sub(r'', text)
# Here text can be any string
df1['text'] = df1['text'].apply(remove_url)

# import string
# exclude = string.punctuation
# def remove_punc1(text):
# 	return text.translate(str.maketrans('', '', exclude))
# df1['text'] = df1['text'].apply(remove_punc1)

def chat_conversion(text1):
  new_text = []
  for text in text1.strip().split('\n'):
    new_text1 = []
    for w in text.split():
      if w.upper() in chat_words:
        new_text1.append(chat_words[w.upper()])
      else:
        new_text1.append(w)
    new_text1 = " ".join(new_text1)
    new_text.append(new_text1)
  return "\n".join(new_text)
df1['text'] = df1['text'].apply(chat_conversion)
```

```python
def replace_isolated_dots(text):
    text = text.replace('..', '.')
    # Find all positions of full stops in the text
    dot_indices = [m.start() for m in re.finditer(r'\.', text)]
    text_list = list(text)  # Convert to list for easy in-place editing

    for i in dot_indices:
        start = max(0, i - 5)
        end = min(len(text), i + 6)  # +6 to include the 5 characters after the dot
        neighborhood = text[start:i] + text[i+1:end]  # exclude the dot itself

        if '.' not in neighborhood:
            text_list[i] = '\n'  # Replace dot with newline

    return ''.join(text_list)

def replace_end_punc(line):
    line = line.replace('।', '\n')
    line = line.replace('?', '?\n')
    line = line.replace('!', '!\n')
    line = line.replace('\n\n', '\n')
    return line

df1['text'] = df1['text'].apply(replace_isolated_dots).apply(replace_end_punc)
```

```python
# Detect language safely
def detect_lang(text):
    try:
        lang = detect(text)
        return lang
    except:
        return None
# Check if sentence is in Devanagari script
def is_devanagari(text):
    result = bool(re.search(r'\p{Devanagari}', text))
    return result

# Check if sentence is in Latin (Roman) script
def is_latin(text):
    result = bool(re.search(r'\p{Latin}', text))
    return result
# Filter function
def keep_sentence(text):
    # lang = detect_lang(text)
    # if lang not in ['en', 'hi']:
    #     return False

    has_devanagari = is_devanagari(text)
    has_latin = is_latin(text)
    return has_devanagari or has_latin

def clean_text_block(text):
    lines = text.strip().split('\n')
    filtered = [line for line in lines if keep_sentence(line)]
    return '\n'.join(filtered)
```

```python
df1['text'] = df1['text'].apply(clean_text_block)
df1 = df1[df1['text'].str.strip() != '']
df1.reset_index(drop=True, inplace=True)
```

```python
def transliterate_to_devanagari(text):
    return transliterate(text, ITRANS, DEVANAGARI)

def transliterate_text_block(text):
    lines = text.strip().split('\n')
    new_lines = []
    for sent in lines:
        collection = sent.strip().split()
        for word in collection:
            counter = False
            if(is_devanagari(word)):
                counter = True
                break
        if(counter):
            new_collection = []
            for word in collection:
                if(is_latin(word) & (word.upper() != word)):
                    word = transliterate_to_devanagari(word)
                new_collection.append(word)
            new_collection = " ".join(new_collection)
            new_lines.append(new_collection)
        else:
            new_lines.append(sent)
    return "\n".join(new_lines)

df1['text'] = df1['text'].apply(transliterate_text_block)
```

```python
from IPython.display import display, HTML
import difflib

def highlight_diff_html(str1, str2):
    d = difflib.Differ()
    diff = list(d.compare(str1, str2))

    result_html = ""
    for char in diff:
        if char.startswith('  '):
            result_html += char[2:]  # same character
        elif char.startswith('- '):
            result_html += f"<span style='color:red;font-weight:bold'>[-{char[2:]}]</span>"
        elif char.startswith('+ '):
            result_html += f"<span style='color:green;font-weight:bold'>[+{char[2:]}]</span>"

    display(HTML(f"<div style='font-family:monospace;font-size:16px'>{result_html}</div>"))
```

```python
from textblob import TextBlob
for para in df1['text']:
  textBlb = TextBlob(para)
  para1 = textBlb.correct().string
  if(para1 != para):
    highlight_diff_html(para, para1)
```
